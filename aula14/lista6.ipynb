```python
#6A
from datetime import datetime
import json
import os

class Cliente:
    def __init__(self, id, nome, email, telefone):
        self.set_id(id)
        self.set_nome(nome)
        self.set_email(email)
        self.set_telefone(telefone)

    def set_id(self,id):
        if id < 0: raise ValueError('Id inválido')  
        else: self.__id = id
    def get_id(self):
        return self.__id
    
    def set_nome(self,nome):
        if nome == '': raise ValueError('Nome inválido')
        else: self.__nome = nome
    def get_nome(self):
        return self.__nome
    
    def set_email(self,email):
        if email == '': raise ValueError('Email inválido')
        else: self.__email = email
    def get_email(self):
        return self.__email
    
    def set_telefone(self,telefone):
        if telefone == '': raise ValueError('Telefone inválido')
        else: self.__telefone = telefone
    def get_telefone(self):
        return self.__telefone
    
    def to_dict(self):
        return{
            'id' : self.get_id(),
            'nome' : self.get_nome(),
            'email' : self.get_email(),
            'telefone' : self.get_telefone(),
        }
    
    @staticmethod
    def from_dict(dicionario):
        return Cliente(
            dicionario['id'],
            dicionario['nome'],
            dicionario['email'],
            dicionario['telefone'],
        )
    
    def __str__(self):
        return f'Id - {self.__id} | Nome - {self.__nome} | Email - {self.__email} | Telefone - {self.__telefone}'
    
class ClienteUI:
    __objetos = []

    @staticmethod
    def main():
        escolha = 0
        while escolha != 8:
            escolha = ClienteUI.menu()
            if escolha == 1: ClienteUI.inserir()
            if escolha == 2: ClienteUI.listar()
            if escolha == 3: ClienteUI.listar_id()
            if escolha == 4: ClienteUI.atualizar()
            if escolha == 5: ClienteUI.excluir()
            if escolha == 6: ClienteUI.abrir()
            if escolha == 7: ClienteUI.salvar()

    @staticmethod
    def menu():
        pergunta = int(input('1 - Inserir | 2 - Listar | 3 - Listar Id | 4 - Atualizar | 5 - Excluir | 6 - Abrir | 7 - Salvar | 8 - Fim'))    
        return pergunta

    @staticmethod   
    def inserir():
        id = int(input('Informe o id:'))
        nome = input('Informe o nome:')
        email = input('Informe o email:')
        telefone = int(input('Informe o telefone:'))
        c = Cliente(id, nome, email, telefone)
        ClienteUI.__objetos.append(c)

    @staticmethod
    def listar():
        if len(ClienteUI.__objetos) == 0:
            print('Nenhum cliente cadastrado')
        for p in ClienteUI.__objetos:
            print(p)

    @staticmethod
    def listar_id():
        id = int(input('Informe o id:'))
        for p in ClienteUI.__objetos:
            if p.get_id() == id:
                print(p)
                return
        print('Id inexistente')

    @staticmethod
    def atualizar():
        id = int(input('Informe o id para atualizar:'))
        for p in ClienteUI.__objetos:
            if p.get_id() == id:
                nome = input('Informe o nome:')
                email = input('Informe o email:')
                telefone = int(input('Informe o telefone:'))
                ClienteUI.__objetos.remove(p)
                novo = Cliente(id, nome, email, telefone)
                ClienteUI.__objetos.append(novo)
                return 
        print('Esse cliente não existe')
    
    @staticmethod
    def excluir():
        id = int(input('Informe o id para excluir:'))
        for p in ClienteUI.__objetos:
            if p.get_id() == id:
                ClienteUI.__objetos.remove(p)
                return 
        print('Esse cliente não existe')

    @staticmethod
    def abrir():
        if not os.path.exists("clientes.json"):
            print("Arquivo de clientes não encontrado.")
        else:
            with open("clientes.json", "r") as arquivo:
                lista = json.load(arquivo)
                ClienteUI.__objetos = [Cliente.from_dict(d) for d in lista]
            print("Clientes carregados com sucesso.")

    @staticmethod
    def salvar():
        lista = [c.to_dict() for c in ClienteUI.__objetos]
        with open("clientes.json", "w") as arquivo:
            json.dump(lista, arquivo)
        print("Clientes salvos com sucesso.")

ClienteUI.main()
```

```python
#6B
from datetime import datetime
import json
import os

class Contato:
    def __init__(self, id, nome, email, telefone, dn):
        self.set_id(id)
        self.set_nome(nome)
        self.set_email(email)
        self.set_telefone(telefone)
        self.set_dn(dn)

    def set_id(self,id):
        if id < 0: raise ValueError('Id inválido')  
        else: self.__id = id
    def get_id(self):
        return self.__id
    
    def set_nome(self,nome):
        if nome == '': raise ValueError('Nome inválido')
        else: self.__nome = nome
    def get_nome(self):
        return self.__nome
    
    def set_email(self,email):
        if email == '': raise ValueError('Email inválido')
        else: self.__email = email
    def get_email(self):
        return self.__email
    
    def set_telefone(self,telefone):
        if telefone == '': raise ValueError('Telefone inválido')
        else: self.__telefone = telefone
    def get_telefone(self):
        return self.__telefone
    
    def set_dn(self,dn):
        if dn > datetime.today(): raise ValueError('Data de nascimento inválida')
        else: self.__dn = dn
    def get_dn(self):
        return self.__dn
    
    def to_dict(self):
        return{
            'id' : self.get_id(),
            'nome' : self.get_nome(),
            'email' : self.get_email(),
            'telefone' : self.get_telefone(),
            'dn' : self.get_dn().strftime('%d/%m/%Y')
        }
    
    @staticmethod
    def from_dict(dicionario):
        return Contato(
            dicionario['id'],
            dicionario['nome'],
            dicionario['email'],
            dicionario['telefone'],
            datetime.strptime(dicionario['dn'], '%d/%m/%Y')
        )

    def __str__(self):
        return f'ID - {self.__id} | Nome - {self.__nome} | E-mail - {self.__email} | Telefone - {self.__telefone} | Data de Nascimento - {self.__dn}'
    
class ContatoUI:
    __contatos =[]

    @staticmethod
    def main():
        escolha = 0
        while escolha != 10:
            escolha = ContatoUI.menu()
            if escolha == 1: ContatoUI.inserir()
            if escolha == 2: ContatoUI.listar()
            if escolha == 3: ContatoUI.listar_Id()
            if escolha == 4: ContatoUI.atualizar()
            if escolha == 5: ContatoUI.excluir()
            if escolha == 6: ContatoUI.pesquisar()
            if escolha == 7: ContatoUI.aniversariantes()
            if escolha == 8: ContatoUI.abrir()
            if escolha == 9: ContatoUI.salvar()

    @staticmethod
    def menu():
        pergunta = int(input('1 - Inserir | 2 - Listar | 3 - Listar Id | 4 - Atualizar | 5 - Excluir | 6 - Pesquisar | 7 - Aniversariantes | 8 - Abrir | 9 - Salvar | 10 - Fim'))    
        return pergunta

    @staticmethod   
    def inserir():
        id = int(input('Informe o id:'))
        nome = input('Informe o nome:')
        email = input('Informe o email:')
        telefone = int(input('Informe o telefone:'))
        data_nascimento = datetime.strptime(input('Informe a data de nascimento:'), '%d/%m/%Y')
        c = Contato(id, nome, email, telefone, data_nascimento)
        ContatoUI.__contatos.append(c)

    @staticmethod
    def listar():
        if len(ContatoUI.__contatos) == 0:
            print('Nenhum contato cadastrado')
        for p in ContatoUI.__contatos:
            print(p)

    @staticmethod
    def listarId():
        id = int(input('digite o id'))
        for c in ContatoUI.__contatos:
            if c.get_id() == id:
                print(c)
                return 
        print('Id inexistente')

    @staticmethod
    def atualizar():
        id = int(input('Informe o id para atualizar:'))
        for p in ContatoUI.__contatos:
            if p.get_id() == id:
                nome = input('Informe o nome:')
                email = input('Informe o email:')
                telefone = int(input('Informe o telefone:'))
                data_nascimento = datetime.strptime(input('Informe a data de nascimento:'), '%d/%m/%Y')
                ContatoUI.__contatos.remove(p)
                novo = Contato(id, nome, email, telefone, data_nascimento)
                ContatoUI.__contatos.append(novo)
                return 
        print('Esse contato não existe')
    
    @staticmethod
    def excluir():
        id = int(input('Informe o id para excluir:'))
        for p in ContatoUI.__contatos:
            if p.get_id() == id:
                ContatoUI.__contatos.remove(p)
                return 
        print('Esse contato não existe')

    @staticmethod
    def pesquisar():
        nome = input("Nome do contato para pesquisar: ")
        for p in ContatoUI.__contatos:
            if p.get_nome().startswith(nome):
                print(p)

    @staticmethod
    def aniversariantes():
        mes = int(input('Digite o número do mês: '))
        for p in ContatoUI.__contatos:
            if p.get_dn().month == mes:
                print(p)

    @staticmethod
    def abrir():
        if not os.path.exists("contatos.json"):
            print("Arquivo de contatos não encontrado.")
        else:
            with open("contatos.json", "r") as lista_contatos:
                lista_dict = json.load(lista_contatos)
                ContatoUI.__contatos = [Contato.from_dict(d) for d in lista_dict]
            print("Contatos carregados com sucesso.")

    @staticmethod
    def salvar():
        lista_dict = [c.to_dict() for c in ContatoUI.__contatos]
        with open("contatos.json", "w") as lista_contatos:
            json.dump(lista_dict, lista_contatos)
        print("Contatos salvos com sucesso.")

ContatoUI.main()
```
