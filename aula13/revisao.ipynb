```python
#questão 1
from datetime import datetime
from datetime import timedelta

class Treino:
    def __init__(self, id, data, distancia, tempo):
        self.set_id(id)
        self.set_data(data)
        self.set_distancia(distancia)
        self.set_tempo(tempo)

    def set_id(self,id):
        if id < 0: raise ValueError('Id inválido')
        else: self.__id = id
    def get_id(self):
        return self.__id
    
    def set_data(self,data):
        if data > datetime.today(): raise ValueError('Data inválida')
        else: self.__data = data
    def get_data(self):
        return self.__data
    
    def set_distancia(self,distancia):
        if distancia < 0: raise ValueError('Distância inválida')
        else: self.__distancia = distancia
    def get_distancia(self):
        return self.__distancia
    
    def set_tempo(self, tempo): 
        if tempo < timedelta(0): raise ValueError('Tempo inválido')
        else: self.__tempo = tempo
    def get_tempo(self):
        return self.__tempo
    
    def __str__(self):
        return f'ID - {self.__id} | DATA - {self.__data} | DISTÂNCIA PERCORRIDA - {self.__distancia} | TEMPO DA CORRIDA - {self.__tempo}'
    
class TreinoUI:
    __treino = []

    @classmethod
    def main(cls):
        escolha = 0
        while escolha != 7:
            escolha = TreinoUI.menu()
            if escolha == 1: TreinoUI.Inserir()
            if escolha == 2: TreinoUI.Listar()
            if escolha == 3: TreinoUI.ListarId()
            if escolha == 4: TreinoUI.Atualizar()
            if escolha == 5: TreinoUI.Excluir()
            if escolha == 6: TreinoUI.MaisRapido()
            
    @classmethod
    def menu(cls):
        pergunta = int(input('1 - INSERIR | 2 - LISTAR TODOS | 3 - LISTAR ID | 4 - ATUALIZAR | 5 - EXCLUIR | 6 - MAIS RÁPIDO | 7 -FIM'))
        return pergunta
    
    @classmethod
    def Inserir(cls):
        id = int(input('ID: '))
        data = datetime.strptime(input('DATA: '), '%d/%m/%Y')
        distancia = int(input('DISTÂNCIA: '))
        h,m,s = map(int, input('TEMPO (hh:mm:ss)').split(':'))
        tempo = timedelta(hours=h, minutes=m, seconds=s)
        c = Treino(id, data, distancia, tempo)
        cls.__treino.append(c)

    @classmethod
    def Listar(cls):
        if len(cls.__treino) == 0:
            print('Nenhum contato cadastrado')
        else:
            for c in cls.__treino:
                print(c)

    @classmethod
    def ListarId(cls):
        id = int(input('Insira o id para listar: '))
        for c in cls.__treino:
            if c.get_id() == id:
                print(c)
    
    @classmethod
    def Atualizar(cls):
        id = int(input('Insira o id para atualizar: '))
        for c in cls.__treino:
            if c.get_id() == id:
                data = datetime.strptime(input('DATA: '), '%d/%m/%Y')
                distancia = int(input('DISTÂNCIA: '))
                h,m,s = map(int, input('TEMPO (hh:mm:ss)').split(':'))
                tempo = timedelta(hours=h, minutes=m, seconds=s)
                cls.__treino.remove(c)
                novo = Treino(id, data, distancia, tempo)
                cls.__treino.append(novo)
                
    @classmethod
    def Excluir(cls):
        id = int(input('Insira o id para atualizar: '))
        for c in cls.__treino:
            if c.get_id() == id:
                cls.__treino.remove(c)

    @classmethod
    def MaisRapido(cls):
        maior = cls.__treino[0]
        for c in cls.__treino:
            if c.get_distancia() / c.get_tempo().seconds > maior.get_distancia() / maior.get_tempo().seconds:
                maior = c
        print(maior)
               
TreinoUI.main()
```
